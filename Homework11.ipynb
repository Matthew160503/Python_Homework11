```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')
fx = -12*x**4 - 18*x**3 + 5*x**2 + 10*x - 30
print(fx)
```

```python
# Построение графика
p = plot(fx,(x,-100,100))
```

```python
# Определение корней
print(solve(fx))
```

```python
# Вычислить вершину
appex = solve(diff(fx))
print(appex)
```

```python
# Найти интервалы, на которых функция возрастает
solve(diff(fx) > 0)
```

```python
# Найти интервалы, на которых функция убывает
solve(diff(fx) < 0)
```

```python
# Определить промежутки, на котором f > 0
solve(fx > 0)
```

```python
# Определить промежутки, на котором f < 0
solve(fx < 0)
```
